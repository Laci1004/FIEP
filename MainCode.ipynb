```python
from datetime import datetime, timedelta
import pandas as pd
import yfinance as yf
```

```python
file_path = "/content/output_short.csv"
df = pd.read_csv(file_path)
```

```python
df
```

```python
def convert_date_format(date_str):
    return datetime.strptime(date_str, '%m/%d/%Y').strftime('%Y-%m-%d')

# Apply the date format conversion to your DataFrame
df['TransactionDate'] = df['TransactionDate'].apply(convert_date_format)
```

```python
df
```

```python
def get_closing_price(ticker, date):
    # Convert string date to datetime object
    start_date = datetime.strptime(date, '%Y-%m-%d')

    # Add one day to the start date for the end date
    end_date = start_date + timedelta(days=1)

    # Convert back to string format for yfinance
    start_date_str = start_date.strftime('%Y-%m-%d')
    end_date_str = end_date.strftime('%Y-%m-%d')

    # Fetch data from yfinance
    data = yf.download(ticker, start=start_date_str, end=end_date_str)
    if not data.empty and 'Close' in data.columns:
        return data['Close'][0]
    else:
        return None


# Create a new column for the estimated number of shares
df['EstimatedShares'] = None

for index, row in df.iterrows():
    closing_price = get_closing_price(row['Ticker'], row['TransactionDate'])
    if closing_price is not None and closing_price != 0:
        df.at[index, 'EstimatedShares'] = row['Amount'] / closing_price
```

```python
df.to_csv('filename.csv', index=False)
```
